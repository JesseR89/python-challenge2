```python
#modules to import
import os
import csv


#File to Load
csvpath = os.path.join('..','Resources', "budget_data.csv")
analysis_output = os.path.join("Analysis", "budget_output.txt")


#change in months variable 
changes = []
greatest_increase = ['', 0]
greatest_decrease = ['', 10000000]
total_change = []
total_amount = 0 
total_months= 0
```

```python

with open(csvpath) as csvfile:
    csvreader = csv.reader(csvfile)
    header = next(csvreader)
    
    first_row = next(csvreader)
    column_index = header.index('Date')
    money_index = header.index('Profit/Losses')
    total_months = 0
    previous_profit = 0
    total_change = 0
    count = 0
    money = []
    profit_change = 0
    greatest_decrease = float('inf')
    greatest_decrease_date = ''
    #calculation in the changes in "profit/loss"
    for row in csvreader:
        if row[money_index]:
            profit = int(row[1])
            change = profit - previous_profit
            total_change += change
            previous_profit = profit
            count += 1
            average_change = total_change/(count)
        if row[column_index]:
            total_months += 1
            print(f'Total Months: {total_months}')
            print(f'Average change in profit/loss: {average_change}')   
            
        money.append(row)
        greatest_increase = float('inf')
        greatest_increase_date = ""
        #calculate the changes in profit; find row with the greatest increase in profits    
        for i in range(len(money)-1):
            current_profit = int(money[i][1])
            next_profit = int(money[i+1][1])
            profit_change = next_profit - current_profit
        
        if profit_change > greatest_increase:
            greatest_increase = profit_change
            greatest_increase_date = money[i+1][0]
            print(greatest_increase)
            print(greatest_increase_date)
            
       #calculate the greatest decrease       
        date = row[0]
        profit_loss = float(row[1])
        
        if profit_loss < greatest_decrease:
            greatest_decrease = profit_loss
            greatest_decrease_date = date
            
            print(f"The greatest decrease in profits occurred on {greatest_decrease_date} with an amount of ${greatest_decrease}")
```
